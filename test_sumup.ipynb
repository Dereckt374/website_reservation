```python
from sumup import Sumup
from sumup.checkouts import CreateCheckoutBody, ProcessCheckoutBody
import uuid


import os
from dotenv import load_dotenv
load_dotenv(dotenv_path = '.venv/.env')
```

```python
sumpup_api_key = os.getenv("sumup_api_key")
client = Sumup(api_key=sumpup_api_key)
```

```python
# Get merchant profile
merchant = client.merchant.get()
merchant_code = merchant.merchant_profile.merchant_code
print(f"Merchant: {merchant_code}")
```

```python
# Create a checkout
checkout = client.checkouts.create(
    body=CreateCheckoutBody(
        amount=1.00,
        currency="EUR",
        checkout_reference=str(uuid.uuid4()),
        merchant_code=merchant_code,
        description="Test payment",
        redirect_url="https://example.com/success",
        return_url="https://example.com/webhook"
    )
)

print(f"Checkout ID: {checkout.id}")
print(f"Checkout Reference: {checkout.checkout_reference}")
```

```python
client.checkouts.get(checkout.id)
```

```python
client.checkouts.list()
```

```python
client.checkouts.list_available_payment_methods(merchant_code)
```

```python
pro = ProcessCheckoutBody(payment_type="card")
```

```python
client.checkouts.process(id=checkout.id, body=pro)
```

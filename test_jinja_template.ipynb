```python
import os
from datetime import datetime
from dotenv import load_dotenv
load_dotenv(dotenv_path = '.venv/.env')
from jinja2 import Environment, FileSystemLoader
from weasyprint import HTML, CSS
```

```python
path_statics = "reservations\static"
```

```python
# 1. Charger le dossier contenant les templates
env = Environment(loader=FileSystemLoader("reservations/templates"))

# 2. Charger le template HTML
template = env.get_template("mail_template.html")

# 3. Remplir le template avec des données
data = {
    "text1": "Texte petit",
    "text2": "Texte normal",
    "titre_principal": "Titre principal",
    "sous_titre": "Sous-titre en italique",
    "paragraphe1": "Paragraphe 1 avec variables...",
    "titre2": "Titre secondaire",
    "paragraphe2": "Paragraphe 2 avec variables..."
    }


html_out = template.render(data)
```

```python
with open("output.html", "w", encoding="utf-8") as f:
    f.write(html_out)

print("✅ Fichier 'output.html' généré avec succès !")
```

```python
# 4. Générer le PDF avec la feuille de style
HTML(string=html_out).write_pdf(
    "bon_de_reservation.pdf",
    stylesheets=[CSS("style.css")]
)

print("✅ Rapport généré")
```
